# LINEAR REGRESSION

## importing mdules

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

## Importing data

```python

stud_data = pd.read_csv("http://bit.ly/w-data")
print("\n THE GIVEN STUDENT DATA IS\n ")
print(stud_data)
```

## ploting graph for predicting the percentage of a student based on the no.of study hours

```python

stud_data.plot(x='Hours', y='Scores', style='h-',color='maroon')  
plt.title('\n predicting the percentage of a student based on the no.of study hours\n')  
plt.xlabel('Hours')  
plt.ylabel('Percentage Score')  
plt.show()
```

From above graph we predict that by increasing the study hours a student can incresen their percentage scores

### Assigning attributes and lables  values to variables 

```python
X = stud_data.iloc[:, :-1].values  
y = stud_data.iloc[:, 1].values 
```

### Spliting  data into training and testing

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0)
```

## TRAINING ALGORITHM

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

print("\n Training complete.\n ")
```

## Plotting the regression line

```python

data_line = regressor.coef_*X+regressor.intercept_
plt.scatter(X, y)
plt.plot(X, data_line, color='brown',);
plt.show()
```

## Predicting the scores

```python
y_pred = regressor.predict(X_test)
```

```python
data = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
data
```

## Predicting the score of a student who studies for 9.25 

```python
hours = [9.25]
result = regressor.predict([hours])
print("Score = {}".format(round(result[0],5)))
```

## accuracy of the model

```python
print('Mean absolute error: ',mean_absolute_error(y_test,y_pred))
```
